# Question 4
Write s python program to display list of respected former finance minister of India(i.e.
Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm and make
data frame.

```python
page = requests.get('https://presidentofindia.nic.in/former-presidents')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
presidents = []

for i in soup.find_all('div', class_='desc-sec'):
    h3_tag = i.find('h3')
    if h3_tag:
        president_name = h3_tag.text
        presidents.append(president_name)

print(presidents)
```

```python
term_of_office= []

for i in soup.find_all('div', class_='desc-sec'):
    h5_tag = i.find('h5')
    if h5_tag:
        president_term = h5_tag.text
        term_of_office.append(president_term)

print(term_of_office)
```

```python
df = pd.DataFrame({'President_Name':presidents,
                  'Term_Of_Office':term_of_office})
df
```
